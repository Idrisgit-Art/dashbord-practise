```python
import plotly.express as px
import pandas as pd
```

```python
df = pd.read_csv("Impact_of_Remote_Work_on_Mental_Health.csv")
```

```python
df.isnull().sum()
```

```python
df["Mental_Health_Condition"] = df["Mental_Health_Condition"].fillna("Unknown")
df["Physical_Activity"] =  df["Physical_Activity"].fillna("Unknown")
```

```python
df.isnull().sum()
```

```python
fig1 = px.bar(df, x='Mental_Health_Condition', y='Work_Life_Balance_Rating',
    title='Work-Life Balance by Mental Health Condition',
    labels={
        'Mental_Health_Condition': 'Mental Health Condition',
        'Work_Life_Balance_Rating': 'Work-Life Balance Rating'
    },
    hover_data=["Stress_Level", "Number_of_Virtual_Meetings"] 
)
fig1.update_layout(title_x=0.5)
fig1.show()
```

```python
import plotly.express as px

fig2 = px.scatter(
    df,
    x="Satisfaction_with_Remote_Work",
    y="Stress_Level", 
    color="Job_Role",
    size="Hours_Worked_Per_Week",
    title="Stress Level vs. Satisfaction with Remote Work",
    labels={
        "Satisfaction_with_Remote_Work": "Satisfaction with Remote Work",
        "Stress_Level": "Stress Level"
    },
    hover_name="Employee_ID" 
)

fig2.update_layout(title_x=0.5)
fig2.show()
```

```python
fig3 = px.density_heatmap(
    df,
    x="Company_Support_for_Remote_Work",
    y="Social_Isolation_Rating",
    title="Impact of Company Support on Social Isolation",
    labels={
        "Company_Support_for_Remote_Work": "Company Support for Remote Work",
        "Social_Isolation_Rating": "Social Isolation Rating"
    },
    color_continuous_scale="viridis"
)
fig3.update_layout(title_x=0.5)
fig3.show()
```

```python
fig4 = px.pie(df,names="Access_to_Mental_Health_Resources",
            title="Access to Mental Health Resources",
             hole=0.35)
fig4.update_layout(title_x=0.5)
fig4.show()
```

```python
import plotly.express as px
#scatter plot
scatter_fig = px.scatter(
    df,
    x="Satisfaction_with_Remote_Work",
    y="Stress_Level",
    color="Job_Role",
    size="Hours_Worked_Per_Week",
    title="Stress Level vs. Satisfaction with Remote Work",
    labels={
        "Satisfaction_with_Remote_Work": "Satisfaction with Remote Work",
        "Stress_Level": "Stress Level"
    },
    hover_name="Employee_ID"
)

# Heatmap plot
heatmap_fig = px.density_heatmap(
    df,
    x="Company_Support_for_Remote_Work",
    y="Social_Isolation_Rating",
    title="Impact of Company Support on Social Isolation",
    labels={
        "Company_Support_for_Remote_Work": "Company Support for Remote Work",
        "Social_Isolation_Rating": "Social Isolation Rating"
    },
    color_continuous_scale="viridis"
)

# Bar Chart plot
bar_fig = px.bar(
    df,
    x="Mental_Health_Condition",
    y="Work_Life_Balance_Rating",
    title="Work-Life Balance by Mental Health Condition",
    labels={
        "Mental_Health_Condition": "Mental Health Condition",
        "Work_Life_Balance_Rating": "Work-Life Balance Rating"
    },
    hover_data=["Stress_Level", "Number_of_Virtual_Meetings"]
)

# Pie Chart plot
pie_fig = px.pie(
    df,
    names="Access_to_Mental_Health_Resources",
    title="Access to Mental Health Resources",
    hole=0.35
)
```

```python
from plotly.subplots import make_subplots

fig = make_subplots(
    rows=2, cols=2,
    subplot_titles=(
        "Scatter: Stress vs Remote Work Satisfaction",
        "Heatmap: Company Support & Isolation",
        "Bar: Work-Life Balance & Mental Health",
        "Pie: Access to Mental Health Resources"
    ),
    specs=[[{"type": "scatter"}, {"type": "heatmap"}],
           [{"type": "bar"}, {"type": "domain"}]]
)

for trace in scatter_fig.data:
    fig.add_trace(trace, row=1, col=1)

for trace in heatmap_fig.data:
    fig.add_trace(trace, row=1, col=2)

for trace in bar_fig.data:
    fig.add_trace(trace, row=2, col=1)

for trace in pie_fig.data:
    fig.add_trace(trace, row=2, col=2)

fig.update_layout(
    height=900,
    width=1200,
    title_text="Remote Work & Mental Health Dashboard",
    showlegend=False
)

fig.show()
```

```python
fig.write_html("interactive_dashbord.html")
```
